```python
# 1 
import pandas as pd

data_frame_covid = pd.read_csv("https://raw.githubusercontent.com/DataRepo2019/Data-files/master/subset-covid-data.csv")
data_frame_covid.head(10) 
```

```python
# 2 
data_frame_covid.query("cases > 50")
```

```python
# 3
data_frame_covid.groupby('continent')[['cases', 'deaths']].mean().reset_index()
```

```python
# 4 
data_frame_covid['on_million'] = data_frame_covid.apply(lambda row: (row['cases']/row['population']) * 1_000_000, axis=1)
data_frame_covid
```

```python
# 5
data_frame_covid['cases'] = data_frame_covid['cases'].fillna(0)
data_frame_covid
```

```python
# 6
data_frame_covid.sort_values('cases', ascending=False)
```

```python
# 7
data_frame_covid[['country', 'cases', 'deaths']]
```

```python
# 8
data_frame_covid.set_index('country_code', inplace=True)
data_frame_covid
```
